```python
import pandas as pd
import os
from dotenv import load_dotenv
from apify_client import ApifyClient


load_dotenv()
```

```python
api_token = os.getenv("API_TOKEN")

if not api_token:
    raise ValueError("API_TOKEN tidak ditemukan di .env")

client = ApifyClient(api_token)

run_input = {
    "commentsPerPost": 500,
    "excludePinnedPosts": False,
    "maxRepliesPerComment": 5,
    "postURLs": [
        "https://www.tiktok.com/@laila_abidin_/video/7632336145015262482?is_from_webapp=1&sender_device=pc&web_id=7589117181012379143"
    ],
    "resultsPerPage": 100
}

run = client.actor("BDec00yAmCm1QbMEI").call(run_input=run_input)

items = list(client.dataset(run["defaultDatasetId"]).iterate_items())
```

```python
df = pd.DataFrame(items)

df.head()
```

```python
df.isnull().sum()
```

```python
print(f"Jumlah data yang berhasil di scrape: {len(items)}")
```

```python
df.to_csv("../results/tiktok/tiktok_comments_1.csv", index=False, encoding="utf-8-sig")
df.to_excel("../results/tiktok/tiktok_comments_1.xlsx", index=False)

print("Data berhasil disimpan ke CSV dan Excel")
```
