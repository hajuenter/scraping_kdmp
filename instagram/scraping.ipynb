```python
import pandas as pd
import os
from dotenv import load_dotenv
from apify_client import ApifyClient


load_dotenv()
```

```python
api_token = os.getenv("API_TOKEN")

if not api_token:
    raise ValueError("API_TOKEN tidak ditemukan di .env")

client = ApifyClient(api_token)

run_input = {
    "postUrls": [
        "https://www.instagram.com/reel/DWaXiq_EWY9/"
    ],
    "maxCommentsPerPost": 200,
    "sortOrder": "popular",
}

run = client.actor("499mNnuVGkU2S5rh1").call(run_input=run_input)

items = list(client.dataset(run["defaultDatasetId"]).iterate_items())
```

```python
df = pd.DataFrame(items)

df.head()
```

```python
df.isnull().sum()
```

```python
print(f"Jumlah data yang berhasil di scrape: {len(items)}")
```

```python
df.to_csv("../results/instagram/instagram_comments_1.csv", index=False, encoding="utf-8-sig")
df.to_excel("../results/instagram/instagram_comments_1.xlsx", index=False)

print("Data berhasil disimpan ke CSV dan Excel")
```
